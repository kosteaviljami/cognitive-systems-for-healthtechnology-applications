```python
# import libraries
# %pylab inline

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
```

```python
# Location of the dataset
filename = 'http://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.switzerland.data'

# column names for the data
colnames = 'age', 'sex', 'cp', 'tresbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num'

# read data to data frame
df = pd.read_csv(filename,
                 names = colnames,
                 na_values = '?')

df.head()
```

```python
df['age'].plot()
plt.xlabel('Index')
plt.ylabel('Age (Years)')
plt.grid()
plt.show()
```
